```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import pandas as pd
```

```python
R = np.array([240,280,320,360,400,440,480,520,560,600])

Rth = 440

V = np.array([1.77,1.95,2.11,2.26,2.39,2.51,2.62,2.72,2.81,2.89])

P = V**2/R

table1 = pd.DataFrame({'$R(\Omega)$':R,'$V_L(V)$':V,'$P_L=V^2_L/R_L(Watts)$':P})
table1.index += 1

# print(table2)

model = np.poly1d(np.polyfit(R,P,3))
best_fit_R = np.linspace(np.min(R),np.max(R),1000)
best_fit_P = model(best_fit_R)

fig,ax = plt.subplots(figsize=(10, 8), dpi=80)


plt.plot(best_fit_R,best_fit_P)
plt.scatter(R,P)

y_interval = 0.0001
x_interval = 20

plt.vlines(x=best_fit_R[np.argmax(best_fit_P)], ymin=np.min(best_fit_P)-y_interval/5,ymax=np.max(best_fit_P),
           ls='--',color='purple')
plt.hlines(y=np.max(best_fit_P), xmin=np.min(best_fit_R)-x_interval/5,xmax=best_fit_R[np.argmax(best_fit_P)],
           ls='--',color='purple')


plt.text(530,0.0131,'       SCALE\n\n y-axis:\n 1div = 0.0002Watt\n\n x-axis:\n 1div = 4$\Omega$',
         bbox=dict(facecolor='white', alpha=1))


loc_x = plticker.MultipleLocator(base=x_interval) #puts ticks at regular intervals
ax.xaxis.set_major_locator(loc_x)
loc_y = plticker.MultipleLocator(base=y_interval)
ax.yaxis.set_major_locator(loc_y)

plt.axis([np.min(best_fit_R)-x_interval/5,np.max(best_fit_R)+x_interval/5, np.min(best_fit_P)-y_interval/5,
          np.max(best_fit_P)+y_interval/2])

plt.grid(visible=True, which='major', color='k', linestyle='-')
plt.grid(visible=True, which='minor', color='grey', linestyle='-', alpha=0.2)
plt.minorticks_on()

plt.xlabel('$R_L(\Omega)$')
plt.ylabel('$P_L$(Watts)')
plt.grid(True)

plt.savefig('plots/MPT-440.png')
plt.show()

print(best_fit_R[np.argmax(best_fit_P)])
```
